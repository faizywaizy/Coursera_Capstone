# Introduction

There has been an influx in restaurant openings in the Toronto in the past few years. Often, I notice that there are swathes of similar, if not identical, restaurant types in close proximity of each other. I want to validate this observation through analysis of location data thorugh the Foursquare API, and connect that data to the neighbourhoods that make up the Greater Toronto Area.

### The Problem

Is it possible to predict housing prices by the density of restaurant types in a locale? Often, certain cuisines are thought to be 'higher class' than others. More often than not, fine dining is limited to French, Italian, or Japanese. We can correlate the two by analyzing the clustering of restaurant types, and the mean housing prices within the area code.

### The Stakeholders

People with a particular love for food would be interested in having this problem solved, perhaps to live somewhere near their food preferences of choice. People that are looking to buy a home would be interested, if they used the model to choose future housing opportunities by the restaurants currently. Investors looking to start a business might notice high housing prices and a lack of restaurant types in an area as an opportunity for development.

### Why does it matter?

This analysis matters because the degree of correlation will provide another metric in which the stakeholders may use to their advantage.

```python
housing_df = housing_df.drop('Unnamed: 0',axis=1)
# data from outside of toronto is dropped
housing_toronto = housing_df[housing_df['Address'].str.contains('Toronto')]
housing_toronto = housing_toronto[housing_toronto['Price ($)'] > 200000]
housing_toronto = housing_toronto.reset_index(drop=True)
housing_toronto.head()
```

```python
housing_toronto = housing_toronto[housing_toronto['lat'] <= 43.75]
housing_toronto = housing_toronto[housing_toronto['lat'] >= 43.58]
housing_toronto = housing_toronto[housing_toronto['lng'] >= -79.60]
housing_toronto = housing_toronto[housing_toronto['lng'] <= -79.15]
housing_toronto = housing_toronto.reset_index(drop=True)
```

## Generate Map of Data

### using geoviews, holoviews and datashader

```python
import holoviews as hv
import geoviews as gv
import datashader as ds
import dask.dataframe as dd
from cartopy import crs
from IPython.display import display
pd.options.display.max_columns = None
import re
import multiprocessing as mp
import datashader.transfer_functions as tf
from datashader.utils import lnglat_to_meters as webm
```

```python
df = housing_toronto.copy()
df.loc[:, 'lng'], df.loc[:, 'lat'] = webm(df['lng'],df['lat'])
```

```python
y_range_min = df['lat'].quantile(0.01)
y_range_max = df['lat'].quantile(0.99)
x_range_min = df['lng'].quantile(0.01)
x_range_max = df['lng'].quantile(0.99)
```

```python
sw = webm(x_range_min, y_range_min)
ne = webm(x_range_max, y_range_max)
TO = zip(sw, ne)
```

```python
plot_width = int(500)
plot_height = int(500)
background = "black"
export = partial(export_image, background=background, export_path="export")
cm = partial(colormap_select, reverse=(background!="black"))

display(HTML("<style>.container {width:100% !important; } </style>"))
```

```python
dask_df = dd.from_pandas(df, npartitions=mp.cpu_count())
dask_df = dask_df.compute()
```

## As you can see, there is a rough distribution with an increase in density distrubution in the city core

```python
cvs = ds.Canvas(plot_width, plot_height, *TO)
agg = cvs.points(dask_df, 'lng', 'lat')
```

```python
export(tf.shade(agg, cmap = cm(Greys9, 0.9), how='eq_hist'),"sf_biz_white")
```
